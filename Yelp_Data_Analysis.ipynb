# Functions

```python
def ttest_2sample(sample1, sample2, alpha=0.05, alternative='two-sided'):

  # Perform Levene's Test
  stat, pvalue = st.levene(sample1, sample2)
  print('Levene\'s Test:')
  print(f'The test statistic is: {stat}')
  print(f'The p-value is: {pvalue}')

  # Determine if Welch's test or Independent Two Sample Test will be used
  if pvalue < alpha:
    # Set parameter for Welch's Test
    print('\nVariances are significantly different. Performing Welch\'s Test.')
    equal_var = False

  else:
    # Set parameter for Independent Two Sample Test
    print('\nVariances are not significantly different. Performing Independent Two Sample T-test.')
    equal_var = True

  tscore, pvalue = st.ttest_ind(sample1,
                                sample2,
                                equal_var = equal_var,
                                alternative = alternative)
  print(f'The t-score is: {tscore}')
  print(f'The p-value is: {pvalue}')
  p_value_interpreter(pvalue, alpha)
```

```python
def p_value_interpreter(p_value, alpha):
  if p_value < alpha:
    print('Reject the Null Hypothesis')
  else:
    print('Fail to reject the Null Hypothesis')
```

```python
# Libraries
import pandas as pd
import numpy as np
import scipy.stats as st
```

```python
# Load the data
df = pd.read_json('yelp_academic_dataset_business.json', lines=True)
df.head()
```

# Data Prep and Cleaning

```python
# Filter rows where 'categories' columns contains the string 'restaurants'
restaurant_df = df[df['categories'].str.contains('restaurants', case=False, na=False)]
restaurant_df.head()
```

```python
# Find all unique values in attributes

# Iterate over the 'attributes' column and collect all unique keys
unique_attributes = set()
for attributes in df['attributes'].dropna():
  unique_attributes.update(attributes.keys())

# Print unique keys line by line
for attributes in unique_attributes:
  print(attributes)
```

# Hypotheses

**Hypothesis 1**: Restaurants that are open have better ratings that those that are not

**Hypothesis 2**: Restaurants that deliver food have worse ratings

**Hypothesis 3**: Restaurants that allow smoking are less likely to be open

# Testing Hypothesis 1

**Null Hypothesis**: Open restaurants ratings <= closed restaurants ratings

**Alternative Hypothesis**: Open restaurants ratings > closed restaurants ratings

Perform a 1-tailed 2-sample t-test since we have 2 independent samples

```python
# Isolate ratings for open and closed restaurants

open_rest_ratings = restaurant_df[restaurant_df['is_open'] == 1]['stars']
closed_rest_ratings = restaurant_df[restaurant_df['is_open'] == 0]['stars']
print(open_rest_ratings.head())
print(closed_rest_ratings.head())
```

```python
# Look at the mean for each
print(f'Open restaurants mean rating: {open_rest_ratings.mean()}')
print(f'Closed restaurants mean rating: {closed_rest_ratings.mean()}')
```

```python
# Perform t-test
ttest_2sample(open_rest_ratings,
              closed_rest_ratings,
              0.05,
              'greater')
```

**Conclusion**

- Since the p-value is less than the alpha value (0.05), we reject the null hypothesis that open restaurants have a less than or equal rating to closed restaurants.

# Testing Hypothesis 2

**Null Hypothesis**: Restaurants that deliver food have better (or equal) ratings than those that don't

**Alternative Hypothesis**: Restaurants that deliver food have worse ratings than those that don't

Perform a 1-tailed 2-sample t-test since we have 2 independent samples

```python
# Create a copy of the restaurants data frame
df_h2 = restaurant_df.copy()
df_h2.head()
```

```python
# Isolate restaurants that deliver
delivery_filter = df_h2['attributes'].apply(lambda x: isinstance(x, dict) and x.get('RestaurantsDelivery') == 'True')
delivery_rest = df_h2[delivery_filter]
delivery_rest.head()
```

```python
# Isolate restaurants that don't deliver
no_delivery_filter = df_h2['attributes'].apply(lambda x: isinstance(x, dict) and x.get('RestaurantsDelivery') == 'False')
no_delivery_rest = df_h2[no_delivery_filter]
no_delivery_rest.head()
```

```python
# Isolate ratings for restaurants with and without delivery
delivery_rest_ratings = delivery_rest['stars']
no_delivery_rest_ratings = no_delivery_rest['stars']
```

```python
# Find mean rating for restaurants with and without delivery
print(f'The mean rating for restaurants with delivery is: {delivery_rest_ratings.mean()}')
print(f'The mean rating for restaurants without delivery is: {no_delivery_rest_ratings.mean()}')
```

```python
# Perform t-test
ttest_2sample(delivery_rest_ratings,
              no_delivery_rest_ratings,
              0.05,
              'less')
```

**Conclusion**

- Since the p-value is less than the alpha value (0.05), we reject the null hypothesis that restaurants with delivery have better (or equal) rating when compared to restaurants without delivery.

# Testing Hypothesis 3

**Null Hypothesis**: There is no relationship between the variables 'Smoking' and 'is_open'

**Alternative Hypothesis**: Restaurants that allow smoking are less likely to be open than restaurants that don't

We will perform a Chi-square test since we are comparing the proportion of an attribute between two samples

```python
# Create a copy of the restaurants data frame
df_h3 = restaurant_df.copy()
df_h3.head()
```

```python
# Create a column that states True or Falso for the attribute 'smoking'
df_h3['allows_smoking'] = df_h3['attributes'].apply(lambda x: x.get('Smoking') == 'True' if x else False)
df_h3.head()
```

```python
# Create a contingency table to observe the frequency distribution between smoking and is_open
contingency_table = pd.crosstab(df_h3['allows_smoking'], df_h3['is_open'])
print(contingency_table)
```

```python
# Check number of restaurants that allow smoking
print(df_h3['allows_smoking'].sum())
```

**Conclusion**

- There are no restaurants that allow smoking so this hypothesis cannot be tested
